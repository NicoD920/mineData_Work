```python
import pandas as pd

urls = [
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_1.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_2.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_3.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_4.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_5.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_6.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_7.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_8.csv',
    'https://raw.githubusercontent.com/NicoD920/mineData_Work/main/data/parte_9.csv'
]

# Crear una lista para almacenar los DataFrames
dataframes = []

# Leer los archivos CSV y agregar los DataFrames a la lista
for url in urls:
    data = pd.read_csv(url, sep=",", encoding="UTF-8")
    dataframes.append(data)

# Concatenar los DataFrames en uno solo
df = pd.concat(dataframes, ignore_index=True)
df_puro = df

# Imprimir las primeras filas del DataFrame resultante
print(df.head())
```

```python
import pandas as pd

# Obtener niveles únicos y asignar valores numéricos únicos para 'brewery_name'
niveles_unicos_breweryname = df['brewery_name'].unique()
valores_numericos_breweryname = list(range(1, len(niveles_unicos_breweryname) + 1))

# Crear un DataFrame con la correspondencia para 'brewery_name'
correspondencia_df_breweryname = pd.DataFrame({'categoria': niveles_unicos_breweryname, 'valor_numerico': valores_numericos_breweryname})

# Reemplazar la columna categórica por la columna numérica en el DataFrame original para 'brewery_name'
df['brewery_name'] = df['brewery_name'].map(correspondencia_df_breweryname.set_index('categoria')['valor_numerico'])

# Repetir el proceso para 'beer_style'
niveles_unicos_beerstyle = df['beer_style'].unique()
valores_numericos_beerstyle = list(range(1, len(niveles_unicos_beerstyle) + 1))
correspondencia_df_beerstyle = pd.DataFrame({'categoria': niveles_unicos_beerstyle, 'valor_numerico': valores_numericos_beerstyle})
df['beer_style'] = df['beer_style'].map(correspondencia_df_beerstyle.set_index('categoria')['valor_numerico'])

# Repetir el proceso para 'beer_name'
niveles_unicos_beername = df['beer_name'].unique()
valores_numericos_beername = list(range(1, len(niveles_unicos_beername) + 1))
correspondencia_df_beername = pd.DataFrame({'categoria': niveles_unicos_beername, 'valor_numerico': valores_numericos_beername})
df['beer_name'] = df['beer_name'].map(correspondencia_df_beername.set_index('categoria')['valor_numerico'])
```

### Información importante
Tenemos que las variables "brewery_name", "beer_style" y "beer_name" son variables del tipo catergóricas pero en este caso las hemos etiquetado como numéricas para facilitar el análisis. Hemos guardado la información de a que corresponde cada etiqueta para un posterior análisis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Calcula la matriz de correlaciones
correlation_matrix = df.corr()

# Configura el estilo de seaborn
sns.set(style='white')

# Crea un mapa de calor de las correlaciones
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de Correlaciones')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=df, x='review_overall', bins=20, kde=True)
plt.title('Histograma de review_overall')
plt.xlabel('Valor de review_overall')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que 'df' es tu DataFrame y deseas visualizar la relación entre 'review_overall' y 'beer_style'.

plt.figure(figsize=(25, 8))
ax = sns.barplot(data=df, x='beer_style', y='review_overall')
plt.title('Relación entre review_overall y beer_style')
plt.xlabel('Estilo de cerveza (beer_style)')
plt.ylabel('Valor de review_overall')
plt.xticks(rotation=90)

# Obtén el valor máximo de las puntuaciones de revisión
max_review_value = 4

# Agrega una línea horizontal en el valor máximo
ax.axhline(max_review_value, color='red', linestyle='--', label='Valor referencia')
plt.legend()

plt.show()
```
